## Typesafe your code
image.png



One of Python's "advantages" is the ability to typecast data variables. So, the following is a legitimate piece of code in python.


```python
a = 42
print(a, " is the answer to life, the universe and everything, and is of type", type(a))

a = "This is a pretty cool campus you have here."
print("\n",a,"\nType:", type(a))
```

Now, it was seamless to convert from one type of data to another, in this case from integer to string. While this is pretty useful when you are running a Jupyter Notebook and experimenting with your code, it can cause a lot of issues when you are coding up a industry level software.
<br><br>
Imagine a scenario where you need the user to enter an integer as a parameter to your ML workflow. <br> Instead, they enter "abc".
<br><br>
Your brilliantly developed code with the most advanced algorithm is going to "accept" the input, but fail to run, because it encountered an incompatible data type.
<br><br>
**Reason: Python code is not inherently 'typesafe'.**
<br><br>
**Enter Pydantic to the rescue**
<br>
https://docs.pydantic.dev/

```python
# Importing typesafe specific libraries
from pydantic import BaseModel, Field
from typing import Any, Optional,List, Union
from typing import Dict, Union
from typing_extensions import Annotated, Literal
```

#### Typesafe code in action

```python
class test_obj(BaseModel):
    value: int             # Defining the data type for the "value" object as an integer

# Exp 1: Defining object as an integer
a = test_obj(value=42)
print(a.value, " is the answer to life, the universe and everything, and is of type", type(a.value))
```

```python
class test_obj(BaseModel):
    value: int

# Exp 2: Defining object as a string. Using "try/except" because we expect errors to occur.
# Try running the code below without the "try" command yourself, to see the errors.
try:
  a = test_obj(value="abc")
  print("Exp 2: Defining object as a string")
  print(a.value, " is the answer to life, the universe and everything, and is of type", type(a.value))
except:
  print("Exp 2: Defining object as an string")
  print("Incorrect data type used.")
```

```python
# An interesting capability within Pydantic is that if the entered data can be legitimately converted to the expected data type, it will do so automatically.
a = test_obj(value="42")
print("Exp 2: Defining object as a string")
print(a.value, " is the answer to life, the universe and everything, and is of type", type(a.value))
```

**Typesafe code ensures that there are no surprises in the users' input**
